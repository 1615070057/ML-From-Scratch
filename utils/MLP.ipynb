# MLP

```python
import keras
from optimizers import StochasticGradientDescent
from layers import Dense,Dropout,Activation,BatchNormalization,Flatten
from neural_networks import NeuralNetwork
from losses import CrossEntropy
from sklearn import datasets
from metric import accuracy_score
from progress import to_categorical,train_test_split
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def load_dataset(flatten=False):
    (X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

    X_train = X_train.astype(float) / 255.
    X_test = X_test.astype(float) / 255.

    X_train, X_val = X_train[:-10000], X_train[-10000:]
    y_train, y_val = y_train[:-10000], y_train[-10000:]

    if flatten:
        X_train = X_train.reshape([X_train.shape[0], -1])
        X_val = X_val.reshape([X_val.shape[0], -1])
        X_test = X_test.reshape([X_test.shape[0], -1])

    return X_train, y_train, X_val, y_val, X_test, y_test
```

```python
X_train, Y_train, X_val, Y_val, X_test, Y_test = load_dataset(flatten=True)
```

```python
y_train = to_categorical(Y_train)
y_test = to_categorical(Y_test)
```

## 简单平凡网络结果

```python
optimizer = StochasticGradientDescent()
n_samples, n_features = X_train.shape
n_hidden = 32


clf = NeuralNetwork(optimizer=optimizer,
                    loss=CrossEntropy,
                    validation_data=(X_test, y_test))

clf.add(Dense(n_hidden, input_shape=(n_features,)))
clf.add(Activation('relu'))
clf.add(Dense(n_hidden))
clf.add(Activation('relu'))
clf.add(Dense(10))
clf.add(Activation('softmax'))
```

```python
train_err, val_err = clf.fit(X_train, y_train, n_epochs=100, batch_size=32)
```

```python
n = len(train_err)
training, = plt.plot(range(n), train_err, label="Training Error")
validation, = plt.plot(range(n), val_err, label="Validation Error")
plt.legend(handles=[training, validation])
```

```python
_,acc = clf.test_on_batch(X_test,y_test)
acc
```

## 加入dropout后的效果

```python
clf = NeuralNetwork(optimizer=optimizer,
                    loss=CrossEntropy,
                    validation_data=(X_test, y_test))

clf.add(Dense(n_hidden, input_shape=(n_features,)))
clf.add(Activation('relu'))
clf.add(Dropout(0.2))
clf.add(Dense(n_hidden))
clf.add(Activation('relu'))
clf.add(Dropout(0.5))
clf.add(Dense(10))
clf.add(Activation('softmax'))
```

```python
train_err, val_err = clf.fit(X_train, y_train, n_epochs=100, batch_size=32)
```

```python
n = len(train_err)
training, = plt.plot(range(n), train_err, label="Training Error")
validation, = plt.plot(range(n), val_err, label="Validation Error")
plt.legend(handles=[training, validation])
```

```python
_,acc = clf.test_on_batch(X_test,y_test)
acc
```

### 加入batchNormalization后的效果

```python
optimizer = StochasticGradientDescent()
n_samples, n_features = X_train.shape
n_hidden = 32

clf = NeuralNetwork(optimizer=optimizer,
                    loss=CrossEntropy,
                    validation_data=(X_test, y_test))

clf.add(Dense(n_hidden, input_shape=(n_features,)))
clf.add(Activation('relu'))
clf.add(Dropout(0.2))
clf.add(BatchNormalization())
clf.add(Dense(32))
clf.add(Activation('relu'))
clf.add(BatchNormalization())
clf.add(Dropout(0.5))
clf.add(Dense(10))
clf.add(Activation('softmax'))
```

```python
train_err, val_err = clf.fit(X_train, y_train, n_epochs=100, batch_size=32)
```

```python
n = len(train_err)
training, = plt.plot(range(n), train_err, label="Training Error")
validation, = plt.plot(range(n), val_err, label="Validation Error")
plt.legend(handles=[training, validation])
```
